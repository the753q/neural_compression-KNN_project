### GitHub Repo Loader

```python
from pathlib import Path
import os
import shutil
import subprocess

repo_url = "https://github.com/the753q/neural_compression-KNN_project.git"
repo_dir = Path("/content/neural_compression-KNN_project")

if repo_dir.exists():
    print(f"Removing existing directory: {repo_dir}")
    shutil.rmtree(repo_dir)

subprocess.run(["git", "clone", repo_url, str(repo_dir)], check=True)
os.chdir(repo_dir)
print(f"Current working directory: {Path.cwd()}")
```

### Main Code

```python
project_root = Path.cwd().resolve()
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from scripts import retrieve_dataset

datamodule = retrieve_dataset("Kodak24", batch_size=1)
datamodule.prepare_data()
datamodule.setup()

print(f"Number of Kodak images: {len(datamodule.dataset)}")

num_images_to_show = 4
fig, axes = plt.subplots(1, num_images_to_show, figsize=(16, 4))

for index in range(num_images_to_show):
    image = datamodule.dataset[index]
    axes[index].imshow(image)
    axes[index].set_title(f"Kodak24 #{index + 1}")
    axes[index].axis("off")

plt.tight_layout()
plt.show()
```
