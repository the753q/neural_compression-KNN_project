### GitHub Repo Loader

```python
from pathlib import Path
import os
import shutil
import subprocess

repo_url = "https://github.com/the753q/neural_compression-KNN_project.git"
repo_dir = Path("/content/neural_compression-KNN_project")
content_dir = Path("/content")

os.chdir(content_dir)
print(f"Reset working directory to: {Path.cwd()}")

if repo_dir.exists():
    print(f"Removing existing directory: {repo_dir}")
    shutil.rmtree(repo_dir)

try:
    result = subprocess.run(
        ["git", "clone", repo_url, str(repo_dir)],
        check=True,
        text=True,
        capture_output=True,
    )
    if result.stderr.strip():
        print(result.stderr.strip())
except subprocess.CalledProcessError as error:
    stderr_output = (error.stderr or "").strip()
    raise RuntimeError(
        "Git clone failed. Git said:\n"
        f"{stderr_output or 'No stderr output was captured.'}"
    ) from error

os.chdir(repo_dir)
print(f"Current working directory: {Path.cwd()}")
```
