# Kodak24 with PyTorch Lightning

This notebook loads the `Kodak24` dataset through the project's Lightning data module and shows a few sample images with matplotlib.

```python
from scripts import retrieve_dataset
from pathlib import Path
import sys

import matplotlib.pyplot as plt

project_root = Path.cwd().resolve().parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))
```

```python
datamodule = retrieve_dataset("Kodak24", batch_size=1)
datamodule.prepare_data()
datamodule.setup()

print(f"Number of Kodak images: {len(datamodule.dataset)}")
```

```python
num_images_to_show = 4
fig, axes = plt.subplots(1, num_images_to_show, figsize=(16, 4))

for index in range(num_images_to_show):
    image = datamodule.dataset[index]
    axes[index].imshow(image)
    axes[index].set_title(f"Kodak24 #{index + 1}")
    axes[index].axis("off")

plt.tight_layout()
plt.show()
```
